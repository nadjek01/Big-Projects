```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded_files = files.upload()
```

```python
channel_orange_url = '/content/Channel_Orange.csv'
blonde_url = '/content/Blonde.csv'

df_channel_orange = pd.read_csv(channel_orange_url)
df_blonde = pd.read_csv(blonde_url)
```

```python
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
```

```python
# Apply sentiment analysis to "Channel Orange"

df_channel_orange['compound'] = df_channel_orange['Lyrics'].apply(lambda x: sia.polarity_scores(x)['compound'])
df_blonde['compound'] = df_blonde['Lyrics'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
# Data visualization
plt.figure(figsize=(12, 6))
```

```python
# Plot sentiment distribution for "Channel Orange"
plt.hist(df_channel_orange['compound'], bins=30, alpha=0.5, label='Channel Orange', color='orange', edgecolor='black')


plt.title('Sentiment Analysis on Frank Ocean\'s Lyrics')
plt.xlabel('Compound Sentiment Score')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
# Plot sentiment distribution for "Channel Orange"
plt.hist(df_blonde['compound'], bins=30, alpha=0.5, label='Blonde', color='green', edgecolor='black')

plt.title('Sentiment Analysis on Frank Ocean\'s Lyrics')
plt.xlabel('Compound Sentiment Score')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Concatenate the dataframes for "Channel Orange" and "Blonde"
df_combined = pd.concat([df_channel_orange.assign(album='Channel Orange'),
                        df_blonde.assign(album='Blonde')])

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='album', y='compound', data=df_combined)
plt.title('Distribution of Compound Sentiment Scores for Albums')
plt.xlabel('Album')
plt.ylabel('Compound Sentiment Score')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='No.', y='compound', hue='album', data=df_combined, palette={'Channel Orange': 'orange', 'Blonde': 'skyblue'})
plt.title('Sentiment Scores Over Album Time')
plt.xlabel('Track Number')
plt.ylabel('Compound Sentiment Score')
plt.show()
```
